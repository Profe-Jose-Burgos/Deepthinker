```python
# Diseccionamos nuestro dataframe
num_df = df.drop(columns=['Id', 'Education', 'Marital_Status', 'Kidhome', 'Teenhome', 'Dt_Customer', 'Response', 'Complain'])
obj_df = df.loc[:, ['Education', 'Marital_Status']]
bin_df = df.loc[:, ['Kidhome', 'Teenhome', 'Complain']]
label = df['Response']
```

```python
def plot_digit(dataframe):
    col_name = np.array(dataframe.columns)
    bins = dataframe.shape[0]
    for col in col_name:
        fig, axs = plt.subplots()
        axs.hist(dataframe[col],
                 bins=bins,
                 edgecolor='blue',
                 )
        plt.xlabel('Valores')
        plt.ylabel('Respuestas')
        plt.title(col)
        plt.tight_layout()
        plt.show()
```

```python
def plot_digit2(dataframe):
    col_name = np.array(dataframe.columns)
    #bins = dataframe.shape[0]
    for col in col_name:
        fig, axs = plt.subplots()
        sns.set(rc={"figure.figsize": (8, 4)}); np.random.seed(0)
        x = np.random.randn(100)
        ax = sns.distplot(x)
        plt.show()
        plt.xlabel('Valores')
        plt.ylabel('Respuestas')
        plt.title(col)
        plt.tight_layout()
        plt.show()
```

```python
# Utilizaremos SciPy para sacar la distribución normal estandarizada
from scipy.stats import zscore
```

```python
# Sacamos la distribución normal estandarizada de nuestros datos
z_df = zscore(num_df)
z_df.head()
```

```python
# Transformamos el obj data a OneHot Encoding
onehot_df = pd.get_dummies(obj_df)
onehot_df.head()
```

```python
features = pd.concat([z_df, bin_df, onehot_df], axis=1)
features.head()
```

```python
# Librerías para el modelado
import tensorflow as tf
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,
                                                    label,
                                                    test_size=0.2,
                                                    random_state=101)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Creamos nuestro modelo
from tensorflow.python import metrics
from tensorflow.python.ops.math_ops import sigmoid
model = Sequential()
model.add(InputLayer(input_shape=(features.shape[1], ), name='input')) # Checkear :>
model.add(Dense(units=16, activation='sigmoid', name='hidden2'))
model.add(Dense(units=1, activation='sigmoid', name='output'))

model.compile(optimizer=Adam(),
              loss='binary_crossentropy',
              metrics='accuracy')
```

```python
model.summary()
```

```python
# Entrenamiento
history = model.fit(x=x_train, 
          y=y_train, 
          batch_size=100,
          epochs=110,
          verbose=1,
          validation_data=(x_test, y_test))
```

```python
# Colocamos en un dataframe nuestros resultados
hist = pd.DataFrame(history.history)
hist
```

```python
# Para visualizar e interpretar nuestro dataframe más rápido
def visual(dataframe):
    dataframe = dataframe.astype(str) + '%'
    #dataframe = dataframe.astype(int)
    return dataframe
```

```python
hist_loss = hist.loc[:, ['loss', 'val_loss']]
hist_loss = hist_loss*100
hist_loss = hist_loss.astype(int)
visual(hist_loss)
```

```python
hist_acc = hist.loc[:, ['accuracy', 'val_accuracy']]
hist_acc = hist_acc*100
hist_acc = hist_acc.astype(int)
visual(hist_acc)
```

```python
def plot_loss(history_loss):
    plt.figure()
    plt.title('Pérdida')
    plt.plot(history_loss)
    plt.xlabel("Ciclos (Epoch)")
    plt.ylabel("Valor (%)")
    plt.legend(['loss', 'val_loss'])
    plt.tight_layout()
    plt.show()
```

```python
def plot_acc(history_acc):
    plt.figure()
    plt.title('Exactitud')
    plt.plot(history_acc)
    plt.xlabel("Ciclos (Epoch)")
    plt.ylabel("Valor (%)")
    plt.legend(['accuracy', 'val_accuracy'])
    plt.tight_layout()
    plt.show()
```

```python
plot_loss(hist_loss)
plot_acc(hist_acc)
```

```python
y_pred = model.predict(x_test)
y_pred = np.around(y_pred)
```

```python
def matriz(confusion):
    sns.heatmap(confusion/np.sum(confusion), annot=True, 
            fmt='.1%', cmap='Blues')
    plt.title('Matriz de Confusión')
    plt.xlabel("Predicción")
    plt.ylabel("Actual")
    plt.show()
```

```python
confusion = confusion_matrix(y_test, y_pred)
matriz(confusion)
```
