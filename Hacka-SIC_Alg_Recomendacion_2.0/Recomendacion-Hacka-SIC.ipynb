```python
# Gr치fico de distribuci칩n para cada variable num칠rica
# ==============================================================================
# Ajustar n칰mero de subplots en funci칩n del n칰mero de columnas
def grafica_1(dataframe):
    fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(13, 8))
    axes = axes.flat
    columnas_numeric = dataframe.select_dtypes(include=['float64', 'int']).columns
    columnas_numeric = columnas_numeric.drop('nacimiento')
    
    for i, colum in enumerate(columnas_numeric):
        sns.histplot(
            data    = dataframe,
            x       = colum,
            stat    = "count",
            kde     = True,
            color   = (list(plt.rcParams['axes.prop_cycle'])*2)[i]["color"],
            line_kws= {'linewidth': 2},
            alpha   = 0.3,
            ax      = axes[i]
        )
        axes[i].set_title(colum, fontsize = 7, fontweight = "bold")
        axes[i].tick_params(labelsize = 6)
        axes[i].set_xlabel("")
        
        
    fig.tight_layout()
    plt.subplots_adjust(top = 0.9)
    fig.suptitle('Distribuci칩n variables num칠ricas', fontsize = 10, fontweight = "bold");
```

```python
def plot_digit(dataframe):
    col_name = np.array(dataframe.columns)
    bins = dataframe.shape[0]
    for col in col_name:
        fig, axs = plt.subplots()
        axs.hist(dataframe[col],
                 bins=bins,
                 edgecolor='blue',
                 )
        plt.xlabel('Valores')
        plt.ylabel('Respuestas')
        plt.title(col)
        plt.tight_layout()
        plt.show()
```

```python
# Hemos detectado outliers gracias a la gr치fica, ahora graficaremos en base a ello para ajustar
# =====================================================================================================
ingreso = df['ingresos']
media = (ingreso).mean()
std_x = (ingreso).std()*2
edades = df['nacimiento']
media_y = (edades).mean()
std_y = (edades).std()*2

colors = ['blue']*len(ingreso)
for index, x in enumerate(ingreso):
    if abs(x-media) > std_x:
        colors[index] = 'red'
        
for index, x in enumerate(edades):
    if abs(x-media_y) > std_y:
        colors[index] = 'red'

plt.scatter(edades, ingreso, s=100, color=colors)
plt.axhline(media, color='k', linestyle='--')
plt.axvline(media_y, color='k', linestyle='--')

v=media     #y-position of the center
u=media_y    #x-position of the center
b=std_x     #radius on the y-axis
a=std_y    #radius on the x-axis

t = np.linspace(0, 2*pi, 100)
plt.plot( u+a*np.cos(t) , v+b*np.sin(t) )

plt.xlabel('Nacimiento')
plt.ylabel('Ingresos')
sns.set_style('darkgrid')
plt.show()
```

```python
# Diseccionamos nuestro dataframe
num_df_a1 = df_a1.drop(columns=['educaci칩n', 'estado_civil', 'hogar_infantil', 'hogar_adolescente', 'fecha_afiliado', 'respuestas', 'quejas', 'edad'])
obj_df_a1 = df_a1.loc[:, ['educaci칩n', 'estado_civil']]
bin_df_a1 = df_a1.loc[:, ['hogar_infantil', 'hogar_adolescente', 'respuestas']]
label = df_a1['quejas']
```

```python
# Utilizaremos SciPy para sacar la distribuci칩n normal estandarizada
from scipy.stats import zscore
```

```python
z_df = zscore(num_df_a1)
z_df.head()
```

```python
onehot_df = pd.get_dummies(obj_df_a1)
onehot_df.head()
```

```python
features = pd.concat([z_df, bin_df_a1, onehot_df], axis=1)
features.head()
```

```python
# Librer칤as para el modelado
import tensorflow as tf
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,
                                                    label,
                                                    test_size=0.2,
                                                    random_state=101)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Creamos nuestro modelo
from tensorflow.python import metrics
from tensorflow.python.ops.math_ops import sigmoid
model = Sequential()
model.add(InputLayer(input_shape=(features.shape[1], ), name='input')) # Checkear :>
model.add(Dense(units=16, activation='sigmoid', name='hidden2'))
model.add(Dense(units=1, activation='sigmoid', name='output'))

model.compile(optimizer=Adam(),
              loss='binary_crossentropy',
              metrics='accuracy')
```

```python
model.summary()
```

```python
# Entrenamiento
history = model.fit(x=x_train, 
          y=y_train, 
          batch_size=100,
          epochs=110,
          verbose=1,
          validation_data=(x_test, y_test))
```

```python
# Colocamos en un dataframe nuestros resultados
hist = pd.DataFrame(history.history)
hist
```

```python
# Para visualizar e interpretar nuestro dataframe m치s r치pido
def visual(dataframe):
    dataframe = dataframe.astype(str) + '%'
    return dataframe
```

```python
hist_loss = hist.loc[:, ['loss', 'val_loss']]
hist_loss = hist_loss*100
hist_loss = hist_loss.astype(int)
visual(hist_loss)
```

```python
hist_acc = hist.loc[:, ['accuracy', 'val_accuracy']]
hist_acc = hist_acc*100
hist_acc = hist_acc.astype(int)
visual(hist_acc)
```

```python
def plot_loss(history_loss):
    plt.figure()
    plt.title('P칠rdida')
    plt.plot(history_loss)
    plt.xlabel("Ciclos (Epoch)")
    plt.ylabel("Valor (%)")
    plt.legend(['loss', 'val_loss'])
    plt.tight_layout()
    plt.show()
```

```python
def plot_acc(history_acc):
    plt.figure()
    plt.title('Exactitud')
    plt.plot(history_acc)
    plt.xlabel("Ciclos (Epoch)")
    plt.ylabel("Valor (%)")
    plt.legend(['accuracy', 'val_accuracy'])
    plt.tight_layout()
    plt.show()
```

```python
plot_loss(hist_loss)
plot_acc(hist_acc)
```

```python
y_pred = model.predict(x_test)
y_pred = np.around(y_pred)
```

```python
def matriz(confusion):
    sns.heatmap(confusion/np.sum(confusion), annot=True, 
            fmt='.1%', cmap='Blues')
    plt.title('Matriz de Confusi칩n')
    plt.xlabel("Predicci칩n")
    plt.ylabel("Actual")
    plt.show()
```

```python
confusion = confusion_matrix(y_test, y_pred)
matriz(confusion)
```
